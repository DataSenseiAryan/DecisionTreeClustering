```python
feature_dict = {}
for time in ['60','120','180']:
    for column in ['trxn_count_','spend_amt_']:
        for feature in ['poi_typ','location_region','industry_name']:
            print(feature+'_'+column+time)
            feature_dict[feature+'_'+column+time] = df[['card_holder_id',feature,column+time]].pivot_table(index='card_holder_id',columns=feature,values=column+time,aggfunc='sum',fill_value=0,dropna=False).fillna(0).reset_index()
```

```python
for key in feature_dict:
    feature_dict[key].columns = [column + '_' + key if column!='card_holder_id' else column for column in feature_dict[key].columns]
```

```python
dd = df[['card_holder_id','trxn_count_0','trxn_count_30','trxn_count_60','trxn_count_90','trxn_count_120','trxn_count_150','trxn_count_180','trxn_count_90_ahead','spend_amt_0','spend_amt_30','spend_amt_60','spend_amt_90','spend_amt_120','spend_amt_150','spend_amt_180','spend_amt_90_ahead']].groupby('card_holder_id').sum().reset_index()
for key in feature_dict:
    print(key)
    dd = dd.merge(feature_dict[key],on='card_holder_id',how='left',suffixes=['','_'+key.split('_')[-1]])
```

```python
new_cols = []
for column in dd.columns:
    if column.find('trxn_count')!=-1:
        new_cols.append(column.replace('trxn_count','avg'))
        print(new_cols[-1])
        dd[new_cols[-1]] = dd.apply(lambda x: 1 if x[column]==0 else x[column.replace('trxn_count','spend_amt')]//x[column],axis=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
target_col = 'avg_90_ahead'
train_col = [col for col in new_cols if col!=target_col]
```

```python
control_ids = df.loc[df.cohort_name=='control','card_holder_id'].unique()
test_ids = df.loc[df.cohort_name=='test','card_holder_id'].unique()
```

# Approach 1

```python
dd['cohort_name'] = dd.card_holder_id.apply(lambda x: 'control' if x in control_ids else 'test')
```

```python
train = np.nan_to_num(dd.loc[dd.cohort_name=='test',train_col])
ytrain = np.nan_to_num(dd.loc[dd.cohort_name=='test',target_col])
test = np.nan_to_num(dd.loc[dd.cohort_name=='control',train_col])
ytest = np.nan_to_num(dd.loc[dd.cohort_name=='control',target_col])
```

```python
for num in range(5,80):
    dtr = DecisionTreeRegressor(max_features= 'log2',
     max_leaf_nodes= num,
     min_samples_leaf= 3,
     min_samples_split= 2,
     splitter='best'
       )
    dtr.fit(train,ytrain)
    pred = dtr.predict(test)
    print('For max_leaf_nodes = %s' % num)
    print('Mean Squared Error',mean_squared_error(pred,ytest))
    print('Mean Absolute Error',mean_absolute_error(pred,ytest))
    print('Root Mean Squared Error',mean_squared_error(pred,ytest)**0.5)
    print('Clusters = ',np.unique(pred))
    print()
```

```python
dtr = DecisionTreeRegressor(max_features= 'log2',
 max_leaf_nodes= 6,
 min_samples_leaf= 3,
 min_samples_split= 2,
 splitter='best')
dtr.fit(train,ytrain)
pred_test = dtr.predict(test)
pred_train = dtr.predict(train)
```

```python
cluster = {}
num=0
for val in np.unique(pred_test.tolist()+pred_train.tolist()):
    cluster[val] = num
    num = num+1
```

```python
test_control_groups = {}
for ind,val in enumerate(pred_test.tolist()):
    if cluster[val] in test_control_groups:
        if 'control' in test_control_groups[cluster[val]]:
            test_control_groups[cluster[val]]['control'].append(control_ids[ind])
        else:
            test_control_groups[cluster[val]]['control']=[control_ids[ind]]
    else:
        test_control_groups[cluster[val]]={}
        test_control_groups[cluster[val]]['control']=[control_ids[ind]]
```

```python
for ind,val in enumerate(pred_train.tolist()):
    if cluster[val] in test_control_groups:
        if 'test' in test_control_groups[cluster[val]]:
            test_control_groups[cluster[val]]['test'].append(test_ids[ind])
        else:
            test_control_groups[cluster[val]]['test']=[test_ids[ind]]
    else:
        test_control_groups[cluster[val]]={}
        test_control_groups[cluster[val]]['test']=[test_ids[ind]]
```

```python
for key in test_control_groups:
    print('Group %s'% key)
    print(test_control_groups[key].keys())
    for k in test_control_groups[key]:
        print(k,len(test_control_groups[key][k]))
    print()
```

```python
vals = []
for key in test_control_groups:
    for group in ['control','test']:
        if group in test_control_groups[key]:
            for did in test_control_groups[key][group]:
                vals.append([did,group,key])
```
