```python
def get_rules(tree, feature_names, class_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]

    paths = []
    path = []
    
    def recurse(node, path, paths):
        
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            p1, p2 = list(path), list(path)
            p1 += [f"({name} <= {np.round(threshold, 3)})"]
            recurse(tree_.children_left[node], p1, paths)
            p2 += [f"({name} > {np.round(threshold, 3)})"]
            recurse(tree_.children_right[node], p2, paths)
        else:
            path += [(tree_.value[node], tree_.n_node_samples[node])]
            paths += [path]
            
    recurse(0, path, paths)

    # sort by samples count
    samples_count = [p[-1][1] for p in paths]
    ii = list(np.argsort(samples_count))
    paths = [paths[i] for i in reversed(ii)]
    
    rules = []
    for path in paths:
        rule = ""
        for p in path[:-1]:
            if rule != "":
                rule += " & "
            rule += str(p.strip('()'))
        rules += [rule]
        
    return rules
```

```python
X = df.drop('average_rating', axis =1)
y = df['average_rating']
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.tree import _tree
```

```python
def dtc(X,y):
    start_time = datetime.datetime.now()
    
    feats = X.columns
    num = 9
    dtr = DecisionTreeRegressor(max_features= 'auto',
     max_leaf_nodes= num,
     min_samples_leaf= 10,
     min_samples_split= 5,
     splitter='best', random_state=69)
    dtr.fit(X,y)
    clustering_df = X.copy()
    clustering_df['LabelOfCluster'] = -1
    rules = get_rules(dtr, feats, None)
    i=0
    final_df = pd.DataFrame()
    for r in rules:
        #print(r)
        temp_df = clustering_df.query(r)
        temp_df['LabelOfCluster'].replace(-1, i, inplace=True)
        final_df = final_df.append(temp_df)
        i=i+1
    end_time = datetime.datetime.now()
    time_diff = (end_time - start_time)
    return time_diff, final_df
    
    
    
```

```python
from sklearn import cluster

def kmeans(X,n_clusters):
    start_time = datetime.datetime.now()
    k_means = cluster.MiniBatchKMeans(n_clusters=n_clusters,random_state=69)
    k_means.fit(X)
    clustering_df = X.copy()
    clustering_df['LabelOfCluster'] = -1
    labels = k_means.predict(X)
    clustering_df['LabelOfCluster']= labels   
    end_time = datetime.datetime.now()
    time_diff = (end_time - start_time)
    return time_diff
    
```

```python

def Mshift(X , bandwidth):
    start_time = datetime.datetime.now()
    quantile = 0.3
    bandwidth = cluster.estimate_bandwidth(X, quantile=quantile)
    ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
    ms.fit(X)
    labels = ms.predict(X)
    clustering_df = X.copy()
    clustering_df['LabelOfCluster'] = -1
    clustering_df['LabelOfCluster']= labels
    end_time = datetime.datetime.now()
    time_diff = (end_time - start_time)
    return time_diff
    
    
    
```

```python
def dbscan(X,eps=10):
    start_time = datetime.datetime.now()
    dbscan = cluster.DBSCAN(eps=eps)
    #dbscan.fit(X)
    labels = dbscan.fit_predict(X)
    clustering_df = X.copy()
    clustering_df['LabelOfCluster'] = -1
    clustering_df['LabelOfCluster']= labels
    end_time = datetime.datetime.now()
    time_diff = (end_time - start_time)
    return time_diff
```

```python
from sklearn.neighbors import kneighbors_graph


def hierarchical(X, n_clusters):
    connectivity = kneighbors_graph(X, n_neighbors=3, include_self=False)
    start_time = datetime.datetime.now()
    ward = cluster.AgglomerativeClustering(n_clusters=n_clusters, linkage="ward", connectivity=connectivity)
    labels = ward.fit_predict(X)
    clustering_df = X.copy()
    clustering_df['LabelOfCluster'] = -1
    clustering_df['LabelOfCluster']= labels
    end_time = datetime.datetime.now()
    time_diff = end_time - start_time
    return time_diff
    
    
```

```python
from sklearn import cluster, datasets, mixture
import datetime
def gmm(X, n_clusters):
    start_time = datetime.datetime.now()
    gmm = mixture.GaussianMixture(n_components=n_clusters, covariance_type="full", random_state=69)
    gmm.fit(X)
    labels = gmm.predict(X)
    clustering_df= X.copy()
    clustering_df['LabelOfCluster'] = -1
    clustering_df['LabelOfCluster']= labels
    end_time = datetime.datetime.now()
    time_diff = (end_time - start_time)
    return time_diff
    
```

```python
def birch(X, n_clusters):
    start_time = datetime.datetime.now()
    birch = cluster.Birch(n_clusters=n_clusters)
    birch.fit(X)
    labels = birch.fit_predict(X)
    clustering_df = X.copy()
    clustering_df['LabelOfCLuster'] = -1
    clustering_df['LabelOfCluster'] = labels
    end_time = datetime.datetime.now()
    time_diff = end_time- start_time
    return time_diff
```

```python
time_dtc = dtc(X,y)
time_km = kmeans(X,n_clusters=5)
time_ms = Mshift(X, bandwidth=bandwidth)
time_dbscan = dbscan(X,eps=0.8)
time_hier = hierarchical(X,n_clusters=5)
time_gmm = gmm(X,n_clusters=5)
time_birch = birch(X,n_clusters=5)

```

```python
data = [['DTC',time_dtc.microseconds],['Kmeans',time_km.microseconds],['MeanShift',time_ms.microseconds],
        ['DBSCAN',time_dbscan.microseconds],['Hierarchical',time_hier.microseconds], ['GMM',time_gmm.microseconds]
       , ['Birch',time_birch.microseconds]]
addidas = pd.DataFrame(data,columns=['Model','Timeinms'])
```
